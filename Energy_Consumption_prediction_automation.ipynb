```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('/content/Energy_consumption.csv')
dataset.head()
```

```python
dataset = dataset.iloc[:,1:]
dataset.head()
```

```python
train = dataset.sample(frac=0.80, random_state=101).reset_index(drop=True)

test = dataset.drop(train.index).reset_index(drop=True)

print("Training data :" + str(train.shape))
print("Test data :"+str(test.shape))
```

```python
from pycaret.regression import *
```

```python
exp_regression = setup(data=train, target='EnergyConsumption', session_id = 100)
```
